```python
import statsmodels.api as sm
```

```python
predictors = ['WC', 'Analytic', 'Clout', 'Authentic', 'Tone', 'WPS', 'BigWords', 'Dic', 'Linguistic', 'function', 'pronoun', 'ppron']
outcome = 'Retweet_x'

X = sm.add_constant(df1[predictors])

poisson_model = sm.GLM(df1[outcome], X, family=sm.families.Poisson()).fit()

print(poisson_model.summary())
```

```python
neg_binomial_model = sm.GLM(df1[outcome], X, family=sm.families.NegativeBinomial()).fit()

print(neg_binomial_model.summary())
```
