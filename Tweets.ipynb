```python
def modified_gs(A: np.ndarray, columns_indices: list) -> np.ndarray:
    num_vecs = A.shape[0]
    num_dims = len(columns_indices)

    L = np.zeros(num_vecs)
    for i in range(num_vecs):
        L[i] = np.sqrt(np.sum(A[i, columns_indices] ** 2))

    V = A.copy()
    V[:, columns_indices] /= L[:, np.newaxis]

    B = V.copy()
    for j in range(num_dims):
        B[:, columns_indices[j]] /= np.sqrt(np.sum(B[:, columns_indices[j]] ** 2))
        for k in range(j + 1, num_dims):
            B[:, columns_indices[k]] -= np.dot(B[:, columns_indices[j]], V[:, columns_indices[k]]) * B[:, columns_indices[j]]

    return B
```

```python
A = new_df[['len', 'sum_arousal', 'mean_arousal']].to_numpy()
```

```python
column_indices = [0, 1, 2]
```

```python
orthogonalized_array = modified_gs(A, column_indices)
```

```python
for i, column_name in enumerate(['len_orthogonalized', 'sum_arousal_orthogonalized', 'mean_arousal_orthogonalized']):
    new_df[column_name] = orthogonalized_array[:, i]
```

```python
average_response_time = blp.rt.mean()
```

```python
saved_files = [filename for filename in os.listdir('.') if filename.startswith('intermediate_results_')]
if saved_files:

    saved_files.sort()

    last_saved_file = saved_files[-1]
    result_df = pd.read_csv(last_saved_file)

    last_processed_index = result_df.index[-1] + 1
else:

    result_df = pd.DataFrame()

    last_processed_index = 0

for chunk_start in range(last_processed_index, len(new_df), chunk_size):
    chunk_end = min(chunk_start + chunk_size, len(new_df))
    chunk = new_df.iloc[chunk_start:chunk_end]

    chunk['sum_response_time1'] = chunk['Tweets'].apply(lambda x: np.sum([blp.loc[w, 'rt'] for w in x.split() if w in blp.index]) if len([blp.loc[w, 'rt'] for w in x.split() if w in blp.index]) > 0 else average_response_time)

    chunk['mean_response_time1'] = chunk['sum_response_time1'] / chunk['len']

    result_df = pd.concat([result_df, chunk[['sum_response_time1', 'mean_response_time1']]])

    if (chunk_start + chunk_size) % (10 * chunk_size) == 0:
        result_df.to_csv(f'/content/drive/MyDrive/intermediate_results_{chunk_start + chunk_size}.csv', index=False)

result_df.head
```
