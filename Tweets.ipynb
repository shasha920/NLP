```python
df['norm_len'] = df.Tweets.apply(lambda x: len(x.split()))
df.head(2)
```

```python
anew = pd.read_csv('ANEW_EnglishShortened.csv').set_index('Word')
print(anew.shape)
anew.head()
```

```python
df_anew_results = df[['Tweets']].apply(lambda x: np.array(
    [anew.loc[w] for w in x[0].split() if w in anew.index]).sum(axis=0),
    axis=1, result_type='expand')
df_anew_results.columns = ['sum_valence', 'sum_arousal', 'sum_dominance']
df_anew_results.fillna(0, inplace=True)
df_anew_results = pd.concat([df_anew_results,
                            pd.DataFrame(df_anew_results.values/df.norm_len.values[:, np.newaxis],
                                         columns = ['mean_valence', 'mean_arousal', 'mean_dominance'])],
                            axis=1).fillna(0)
df = pd.concat([df, df_anew_results], axis=1)
df.head(2)
```

```python
blp = pd.read_excel('BLP summary data words.xls')
blp.set_index('spelling', inplace=True)
print(blp.shape)
blp.head()
```

```python
average_response_time = blp.rt.mean()
```

```python
df['sum_response_time1'] = df['Tweets'].apply(lambda x: np.array(
    [blp.loc[w, 'rt'] for w in x.split() if w in blp.index]).sum(axis=0) if len([blp.loc[w, 'rt'] for w in x.split() if w in blp.index]) > 0 else average_response_time).fillna(0)
df['mean_response_time1'] = (df['sum_response_time'] / df['len']).fillna(0)
df.head(2)
```
