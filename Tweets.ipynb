```python
df_anew_results = df[['Tweets']].apply(lambda x: np.array(
    [anew.loc[w] for w in x[0].split() if w in anew.index]).sum(axis=0),
    axis=1, result_type='expand')
df_anew_results.columns = ['sum_valence', 'sum_arousal', 'sum_dominance']
df_anew_results.fillna(0, inplace=True)
df_anew_results = pd.concat([df_anew_results,
                            pd.DataFrame(df_anew_results.values/df.norm_len.values[:, np.newaxis],
                                         columns = ['mean_valence', 'mean_arousal', 'mean_dominance'])],
                            axis=1).fillna(0)
df = pd.concat([df, df_anew_results], axis=1)
df.head(2)
```

```python
def modified_gs(A):
    num_vecs = A.shape[1]
    num_dims = A.shape[0]

    L = np.zeros(num_vecs)
    for i in range(num_vecs):
        L[i] = np.sqrt(A[:, i].T @ A[:, i])

    V = A.copy() / L
    B = V.copy()
    for j in range(0, num_vecs):
        B[:, j] = V[:, j] / np.sqrt(V[:, j].T @ V[:, j])
        for k in range(j, num_vecs):
            V[:, k] = V[:, k] - (B[:, j].T @ V[:, k]) * B[:, j]
    return B
```

```python
subset_cols = ['sum_valence', 'sum_arousal', 'sum_dominance']
subset_data = df[subset_cols].to_numpy()
```

```python
orthogonalized_data = modified_gs(subset_data)
```

```python
orthogonalized_df = pd.DataFrame(orthogonalized_data, columns=subset_cols)
```

```python
df= pd.concat([df, orthogonalized_df], axis=1)
```
